```python
# civica_courtdocs

# Connect to MySQL database
cnx = pymysql.connect(user='buspark', password='U@5p1r3!', host='masscourtsplus.org', database='civica_courtdocs')

# Create a cursor object
cursor = cnx.cursor()

# Define query to get table names
table_query = "SHOW TABLES;"

# Execute the query to get table names
cursor.execute(table_query)

# Fetch all table names
tables = cursor.fetchall()

# Iterate over each table
for table in tables:
    table_name = table[0]
    print("Table:", table_name)

    # Define query to get column names for the current table
    column_query = f"DESCRIBE {table_name};"

    # Execute the query to get column names
    cursor.execute(column_query)

    # Fetch column information
    column_info = cursor.fetchall()

    # Extract and print column names
    column_names = [col[0] for col in column_info]
    print("Column Names:", column_names)
    print()  # Add a newline for better readability

# Close cursor/connection
cursor.close()
cnx.close()
# Closing helps decrease loading time, but too much unclosed will damamge connection
```

```python
#Now since we have all the data names, we can extract the correct querries
# Sample Querry 2
cnx = pymysql.connect(user='buspark', password='U@5p1r3!',host='masscourtsplus.org',
database='wp_courtdocs_RESEARCH') #databsesto choose from
                          # civica_courtdocs
                          # wp_courtdocs
                          # wp_courtdocs_NORMALIZED
                          # wp_courtdocs_RESEARCH

#Column Names of analytics1: ['case_number', 'case_status', 'actor', 'party_name', 'action', 'description', 'date_time', 'judge', 'attorney', 'file_date', 'status_date']
cursor = cnx.cursor(pymysql.cursors.DictCursor) # Usually I leave it as cnx.cursor(), but cnx.cursor(pymysql.cursors.DictCursor) allows your column name to be waht you searched for instead of index numbers
                                                #Return results as dictionaries. This ensures that fetchall() returns a list of dictionaries with column names as keys
query =  """
SELECT
    case_number ,
    case_status ,
    description   # From column name of analytics1 we only wanted case_number,case_status, and description
                                # AS names the column name, otherwise it will be named from asscending order

FROM
    analytics1  #taken from table above,  make sure the name of the table matches, some will be a short name some full name
LIMIT 3; #Sets limit in how many rows you want. comment out LIMIT line to show all the rows
"""

%time

cursor.execute(query) # executing query

results = cursor.fetchall() # fetching result

# Close cursor/connection
cursor.close()
cnx.close()
# Closing helps decrease loading time, but too much unclosed will damamge connection
```
